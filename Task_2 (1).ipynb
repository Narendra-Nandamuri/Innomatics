### Problem 1

```python
class Solution:
    def defangIPaddr(self, address: str) -> str:
        return address.replace(".", "[.]")
sol = Solution()

# Test Case 1
input_ip = "1.1.1.1"
output = sol.defangIPaddr(input_ip)
print(f"Input: {input_ip}")
print(f"Output: {output}")

# Test Case 2
input_ip_2 = "255.100.50.0"
output_2 = sol.defangIPaddr(input_ip_2)
print(f"Input: {input_ip_2}")
print(f"Output: {output_2}")
```

### Problem 2

```python
from typing import List

class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        count = 0
        for num in nums:
            if len(str(num)) % 2 == 0:
                count += 1
        return count
sol = Solution()

# Test Case 1
nums1 = [12, 345, 2, 6, 7896]
result1 = sol.findNumbers(nums1)
print(f"Input: {nums1}")
print(f"Output: {result1}")

# Test Case 2
nums2 = [555, 901, 482, 1771]
result2 = sol.findNumbers(nums2)
print(f"Input: {nums2}")
print(f"Output: {result2}")
```

### Problem 3

```python
import collections

class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        pair_count = 0
        counts = {}
        for num in nums:
            if num in counts:
                pair_count += counts[num]
                counts[num] += 1
            else:
                counts[num] = 1

        return pair_count

sol = Solution()

# Test Case 1
nums1 = [1, 2, 3, 1, 1, 3]
output1 = sol.numIdenticalPairs(nums1)
print(f"Input: {nums1}")
print(f"Output: {output1}")

# Test Case 2
nums2 = [1, 1, 1, 1]
output2 = sol.numIdenticalPairs(nums2)
print(f"Input: {nums2}")
print(f"Output: {output2}")
```

### Problem 4

```python
from typing import List

class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        sorted_nums = sorted(nums)
        rank = {}
        for i, num in enumerate(sorted_nums):
            if num not in rank:
                rank[num] = i
        result = []
        for num in nums:
            result.append(rank[num])

        return result
sol = Solution()

# Test Case 1
nums1 = [8, 1, 2, 2, 3]
output1 = sol.smallerNumbersThanCurrent(nums1)
print(f"Input: {nums1}")
print(f"Output: {output1}")

# Test Case 2
nums2 = [6, 5, 4, 8]
output2 = sol.smallerNumbersThanCurrent(nums2)
print(f"Input: {nums2}")
print(f"Output: {output2}")
```

### Problem 5

```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        digits = str(n)

        product_of_digits = 1
        sum_of_digits = 0

        for digit in digits:
            num = int(digit)
            product_of_digits *= num
            sum_of_digits += num

        return product_of_digits - sum_of_digits
sol = Solution()

# Test Case 1
n1 = 234
output1 = sol.subtractProductAndSum(n1)
print(f"Input: {n1}")
print(f"Product: 2 * 3 * 4 = 24")
print(f"Sum: 2 + 3 + 4 = 9")
print(f"Output: {output1}")

# Test Case 2
n2 = 4421
output2 = sol.subtractProductAndSum(n2)
print(f"Input: {n2}")
print(f"Output: {output2}")
```

### Problem 6

```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        result = 0
        for i in range(n):
            current_num = start + 2 * i
            result = result ^ current_num
        return result
sol = Solution()

# Test Case 1
n1 = 5
start1 = 0
output1 = sol.xorOperation(n1, start1)
print(f"n: {n1}, start: {start1}")
print(f"Output: {output1}")

# Test Case 2
n2 = 4
start2 = 3
output2 = sol.xorOperation(n2, start2)
print(f"n: {n2}, start: {start2}")
print(f"Output: {output2}")
```
