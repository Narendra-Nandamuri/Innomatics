### Problem 1

```python
from typing import List

class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        max1 = 0
        max2 = 0

        for num in nums:
            if num > max1:
                max2 = max1
                max1 = num
            elif num > max2:
                max2 = num
        return (max1 - 1) * (max2 - 1)
sol = Solution()

# Test Case 1
nums1 = [3, 4, 5, 2]
output1 = sol.maxProduct(nums1)
print(f"Input: {nums1}")
print(f"Two largest: 5 and 4")
print(f"Calculation: (5-1) * (4-1) = 4 * 3 = 12")
print(f"Output: {output1}")

# Test Case 2
nums2 = [1, 5, 4, 5]
output2 = sol.maxProduct(nums2)
print(f"Input: {nums2}")
print(f"Output: {output2}")
```

### Problem 2

```python
from typing import List

class Solution:
    def numTeams(self, rating: List[int]) -> int:
        n = len(rating)
        total_teams = 0
        for j in range(1, n - 1):
            left_smaller = 0
            left_larger = 0
            for i in range(j):
                if rating[i] < rating[j]:
                    left_smaller += 1
                elif rating[i] > rating[j]:
                    left_larger += 1

            right_smaller = 0
            right_larger = 0
            for k in range(j + 1, n):
                if rating[k] < rating[j]:
                    right_smaller += 1
                elif rating[k] > rating[j]:
                    right_larger += 1
            total_teams += (left_smaller * right_larger) + (left_larger * right_smaller)

        return total_teams
sol = Solution()

# Test Case 1
rating1 = [2, 5, 3, 4, 1]
output1 = sol.numTeams(rating1)
print(f"Input: {rating1}")
print(f"Output: {output1}")

# Test Case 2
rating2 = [2, 1, 3]
output2 = sol.numTeams(rating2)
print(f"Input: {rating2}")
print(f"Output: {output2}")
```

### Problem 3

```python
from typing import List

class Solution:
    def busyStudent(self, startTime: List[int], endTime: List[int], queryTime: int) -> int:
        count = 0

        for s, e in zip(startTime, endTime):
            if s <= queryTime <= e:
                count += 1

        return count
sol = Solution()

# Test Case 1
startTime1 = [1, 2, 3]
endTime1 = [3, 2, 7]
queryTime1 = 4
output1 = sol.busyStudent(startTime1, endTime1, queryTime1)
print(f"Start Times: {startTime1}")
print(f"End Times:   {endTime1}")
print(f"Query Time:  {queryTime1}")
print(f"Output: {output1}")

# Test Case 2
startTime2 = [4]
endTime2 = [4]
queryTime2 = 4
output2 = sol.busyStudent(startTime2, endTime2, queryTime2)
print(f"Start Times: {startTime2}")
print(f"End Times:   {endTime2}")
print(f"Query Time:  {queryTime2}")
print(f"Output: {output2}")
```

### Problem 4

```python
class Solution:
    def numberOfSteps(self, num: int) -> int:
        steps = 0
        while num > 0:
            if num % 2 == 0:
                num = num // 2
            else:
                num -= 1
            steps += 1

        return steps
sol = Solution()

# Test Case 1
num1 = 14
output1 = sol.numberOfSteps(num1)
print(f"Input: {num1}")
print(f"Output: {output1}")

# Test Case 2
num2 = 8
output2 = sol.numberOfSteps(num2)
print(f"Input: {num2}")
print(f"Output: {output2}")
```

### Problem 5

```python
from typing import List

class Solution:
    def countBits(self, n: int) -> List[int]:
        ans = [0] * (n + 1)
        for i in range(1, n + 1):
            ans[i] = ans[i >> 1] + (i & 1)

        return ans
sol = Solution()

# Test Case 1
n1 = 2
output1 = sol.countBits(n1)
print(f"Input: {n1}")
print(f"Output: {output1}")

# Test Case 2
n2 = 5
output2 = sol.countBits(n2)
print(f"Input: {n2}")
print(f"Output: {output2}")
```
